### Analysis

    Data gathered by the mice animal study testing Capomulin against other drug regiments for squamous cell carcinoma (SCC) was compiled. Mouse g989 was removed from data analysis due to the duplications in the data collection. Summary statistics were calculated for each treatment. The average tumor volume for Capomulin was 40.68 mm3 which was significantly lower than the other treatments apart from Ramicane (40.22 mm3). The variance and standard deviation were also much lower for Capomulin and Ramicane compared to other treatment plans. 
	A bar chart was created to display the number of observed timepoints for each treatment. Note: Propriva has less recorded timepoints as g989 was assigned to that regiment and g989’s data is not considered. A pie chart presents the distribution of sex for the test subjects: 51% male and 49% female. 
	Potential outliers were determined both quantitatively and visually with a boxplot for the drug regimens of Capomulin, Ramicane, Infubinol, and Ceftamin. One potential outlier was found within the Infubinol treatment for mouse ID c326 at the final tumor volume of 36.32 mm3. 
	The final charts solely examined the Capomulin treatment. The line chart presents the tumor volume for a single mouse (l509) over the course of 45 days while being treated with Capomulin. After the 20-day mark, the tumor volume significantly decreases. Lastly, a scatter plot was constructed charting the weight (g) of mice on Capmulin vs. the average tumor volume (mm3). A linear regression was performed and showed a strong correlation of 0.84. The correlation is positive because as mice’s’ weight increases, tumor volume increases as well. 
	Overall, Capomulin and Ramicane outperformed the other cancer treatments over the study period. The company’s drug of interest, Capomulin, successfully decreased SCC tumor volume in mice. 

 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "Data/Mouse_metadata.csv"
study_results_path = "Data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice
mice_count = len(merge_df["Mouse ID"].unique())
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_df = merge_df.loc[merge_df.duplicated(subset=["Mouse ID", "Timepoint"])]
duplicate_mice = duplicate_df["Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989_df = merge_df.loc[merge_df["Mouse ID"] == "g989"]
g989_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df.loc[merge_df["Mouse ID"] != "g989"]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count = len(clean_df["Mouse ID"].unique())
mice_count
```

## Summary Statistics

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
drug_summary = clean_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
drug_summary
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
observed_timepoints = clean_df["Drug Regimen"].value_counts()
drug_type = ["Capomulin", "Ramicane", "Ketapril", "Naftisol", "Zoniferol", "Stelasyn", 
             "Placebo", "Infubinol", "Ceftamin", "Propriva"]

plot = observed_timepoints.plot(kind = "bar")
plot.set_xticklabels(drug_type, rotation= "vertical")
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")

plt.show()

```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
plt.bar(drug_type, observed_timepoints, align = "center")
tick_locations = [value for value in drug_type]
plt.xticks(tick_locations, drug_type, rotation = "vertical")
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
labels = ["Male", "Female"]
colors = ["blue", "orange"]

sex_count= clean_df["Sex"].value_counts()
sex_percent = sex_count / 1880 *100

plot = sex_percent.plot(kind = "pie", labels=labels, colors=colors, autopct="%1.1f%%")

plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_df.groupby("Mouse ID")["Timepoint"].max()
last_timepoint.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_df = pd.merge(last_timepoint, clean_df, on=["Mouse ID", "Timepoint"], how="left")
                     
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drugs:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment_df = merged_df.loc[merged_df["Drug Regimen"] == drug]
    final_treatment_df = treatment_df[["Drug Regimen", "Tumor Volume (mm3)"]]
    
    # add subset 
    tumor_vol_data.append(final_treatment_df)
    final_tumor_vol = pd.concat(tumor_vol_data)

    # Determine outliers using upper and lower bounds
    quartiles = final_treatment_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    volumes = final_treatment_df['Tumor Volume (mm3)']
    outlier = []
    for volume in volumes:
        if volume > upper_bound or volume < lower_bound:
            outlier.append(volume)
    
    print(f"{drug}: IQR {iqr} and outlier(s) {outlier}")
    
   
            
     
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

capomulin = final_tumor_vol.loc[final_tumor_vol['Drug Regimen']=='Capomulin']
x1 = capomulin["Tumor Volume (mm3)"]
ramicane = final_tumor_vol.loc[final_tumor_vol['Drug Regimen']=='Ramicane']
x2 = ramicane["Tumor Volume (mm3)"]
infubinol = final_tumor_vol.loc[final_tumor_vol['Drug Regimen']=='Infubinol']
x3 = infubinol["Tumor Volume (mm3)"]
ceftamin = final_tumor_vol.loc[final_tumor_vol['Drug Regimen']=='Ceftamin']
x4 = ceftamin["Tumor Volume (mm3)"]

fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume Across Drug Regimens')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
ax1.boxplot([x1, x2, x3, x4])
plt.xticks([1, 2, 3, 4], drugs)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_l509=clean_df.loc[clean_df["Mouse ID"] == "l509"]
plot_df = mouse_l509[["Timepoint", "Tumor Volume (mm3)"]]
plot_df = plot_df.set_index("Timepoint")
plot_df.plot()
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin"]
weight = capomulin_df.groupby("Mouse ID")["Weight (g)"].mean()
tumor_volume = capomulin_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()

plt.scatter(weight, tumor_volume, marker="o")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
print(f"The correlation between mouse weight and the average tumor volume is {round(st.pearsonr(weight, tumor_volume)[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, tumor_volume)
regress_values = weight * slope + intercept
plt.scatter(weight, tumor_volume)
plt.plot(weight,regress_values,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
